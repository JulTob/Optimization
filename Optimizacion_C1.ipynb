## üó∫Ô∏è El Espacio de Decisi√≥n: Un Plano de Posibilidades

> El Plano de $ùïè‚Éó$ es un _espacio vectorial_.

Imagina por un momento, un vasto lienzo, no de tela, sino de *posibilidades*. Cada variable que puedes controlar, cada "dial" que puedes ajustar en tu problema, es un eje, una dimensi√≥n en este lienzo. Esto es tu **Espacio de Decisi√≥n**.


Cada punto en este espacio representa una combinaci√≥n *√∫nica* de tus decisiones. Es la consideraci√≥n total de todas las elecciones que *podr√≠as* hacer, antes de que la realidad imponga sus l√≠mites.

¬øCu√°ndo lo llamamos **espacio vectorial**? Cuando en √©l, tus decisiones se comportan con una l√≥gica vectorial.

- Puedes "sumar" dos conjuntos de decisiones:
  - $X_1 + X_2 = X_3$
    - Dos viajes se pueden hacer como un solo viaje que vaya directamente al destino.
    - Combinar ingredientes producen un tercer ingrediente, como mezclar huevos y harina para hacer masa.
  
  
  
  
  
  
  
  
  
  
  
  
  - $ùöî¬∑X(x,y) = X(ùöî¬∑x,ùöî¬∑y)$
    - Puedes "multiplicar" tus decisiones por un n√∫mero.
    - Como al moverte el doble en una direcci√≥n te lleva el doble de lejos.
    - Como hacer el doble de producto con el doble de ingredientes.
    





Cada vector representa un posible estado del sistema.

Esta estructura, esta geometr√≠a abstracta, es lo que permite que las matem√°ticas naveguen por este mar de opciones con una precisi√≥n quir√∫rgica.

Es, en esencia, el *tablero* donde se librar√° el juego de la optimizaci√≥n.

## Variables de decisi√≥n
> $ùö°‚Éó$:
>>  $ùö°‚Éó‚àäùïè‚Éó$   
>> $ùö°_i‚ààùïè_i$

Representan las acciones que se pueden tomar, que pertenecen a un dominio $ùïè$






Imagina que est√°s al mando de una m√°quina compleja, un peque√±o universo de tu problema, un videojuego que imita la realidad. Las **variables de decisi√≥n** son los 'diales' y 'palancas', los botones y joysticks, que puedes ajustar, las 'acciones' que puedes tomar. Son las coordenadas en ese 'Espacio de Decisi√≥n' que acabamos de explorar, donde cada ajuste crea un nuevo punto, una nueva posibilidad.

Estas variables, que a menudo agrupamos en un vector $\mathbf{x}$, representan **aquello que est√° bajo tu control**. Por ejemplo, en nuestro dilema de la fruta, ser√≠an $x_1$ (kilos de manzana) y $x_2$ (kilos de pl√°tano). En un problema de producci√≥n, $x_i$ podr√≠a ser la cantidad del producto $i$ a fabricar. Son las magnitudes que, al variar, cambian el estado de tu sistema y, por ende, el resultado.

¬øC√≥mo se eligen? No se 'inventan' al azar. Se *identifican* como los elementos fundamentales que definen las opciones. Si un factor influye en tu objetivo y puedes modificarlo, es una candidata a variable de decisi√≥n. Si no puedes modificarlo, o no afecta a tu objetivo, entonces es una constante o un factor externo. Es crucial que sean magnitudes que *puedes* manipular.

Cada variable $x_i$ tiene un dominio $X_i$, es decir, el rango de valores que puede tomar. A veces es simplemente $\mathbb{R}$ (cualquier n√∫mero real), otras veces son n√∫meros enteros (no puedes fabricar media silla), o quiz√°s solo valores positivos (no puedes producir cantidades negativas). As√≠, el conjunto de todas tus decisiones posibles es un vector $\mathbf{x} \in \mathbb{X}$, donde $\mathbb{X}$ es el producto cartesiano de los dominios individuales de cada variable.

Simple, ¬øverdad? Son solo los botones que puedes pulsar para operar tu sistema.

# Representaci√≥n Cartesiana

## El Espacio Cartesiano: Cada Punto, Una Decisi√≥n

Imagina nuestro dilema frutal como un vasto terreno de elecciones.

Cada punto en este plano $(x_1, x_2)$ representa una *decisi√≥n √∫nica* sobre cu√°ntos kilos de manzanas ($x_1$) y pl√°tanos ($x_2$) podr√≠as comprar. No hay l√≠mites a√∫n, solo posibilidades.

Por ejemplo:

*   El punto **(1, 0)** significa: 1 kilo de manzanas, 0 de pl√°tanos.
*   El punto **(3, 4)** significa: 3 kilos de manzanas, 4 de pl√°tanos.
*   El punto **(0, 5)** significa: 0 kilos de manzanas, 5 de pl√°tanos.

Cada uno de estos puntos es una 'jugada' diferente en tu "partida" por el desayuno. La belleza de este espacio es que puedes visualizar cualquier combinaci√≥n concebible. Antes de pensar en el coste o la capacidad de tu cesta, ¬°todo es posible aqu√≠!

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define some example points (decisions)
points = {
    '(1, 0)': (1, 0), # 1 manzana, 0 pl√°tanos
    '(3, 4)': (3, 4), # 3 manzanas, 4 pl√°tanos
    '(0, 5)': (0, 5), # 0 manzanas, 5 pl√°tanos
    '(2, 2)': (2, 2), # 2 manzanas, 2 pl√°tanos
    '(5, 1)': (5, 1)  # 5 manzanas, 1 pl√°tano
    }

# Create the plot
plt.figure(figsize=(8, 8))

# Plot each point and label it
for label, (x1, x2) in points.items():
    plt.plot(x1, x2, 'o', markersize=8, label=label)
    plt.text(x1 + 0.1, x2 + 0.1, f'{x1} manzanas \n {x2} pl√°tanos', fontsize=9)

# Set plot limits and labels
plt.xlim(-0.5, 6)
plt.ylim(-0.5, 6)
plt.xlabel(r'$x_1$ (Kilos de Manzanas)')
plt.ylabel(r'$x_2$ (Kilos de Pl√°tanos)')
plt.title('Desayunos: Cada Punto es una Posibilidad')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Decisiones Puntuales')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## ‚õî Restricciones.
> Conjunto Factible   
> $ùîΩ ‚äÜ ùïè$

> "Tras d√≠as tallando un caballo para una estatua, se acerc√≥ el ni√±o a Miguelangelo: '¬øC√≥mo sab√≠as que hab√≠a un caballo dentro del marmol?'. `Muy simple`, le contest√≥ el artista. `Cuando quit√© todo lo que sobrava, solo quedaba el caballo.`

> Una restricci√≥n es una proposici√≥n de los valores de ùö°‚Éó.
> $h‚Äã(ùö°‚Éó)?$    
> $h: TRUE/FALSE$

Las **Restricciones**. Hemos explorado el vasto "universo" de nuestras decisiones. Un lienzo ilimitado donde cada punto es una combinaci√≥n, por absurda que parezca, de manzanas y pl√°tanos.

Pero la vida, como la matem√°tica, no es un juego sin reglas. Entran en el tablero las restricciones.

¬øQu√© son? En esencia, los **cortes y l√≠mites** que la realidad impone a nuestro universo de decisi√≥n. No podemos gastar m√°s del dinero que tenemos, ni llevar m√°s de lo que cabe en nuestra cesta, ¬øverdad? Cada una de estas reglas toma una porci√≥n de nuestro espacio y dice que la decisi√≥n no te vale.

De repente, ese lienzo infinito se ve transformado. Las restricciones no eliminan posibilidades al azar; las **esculpen**. Cada "no puedes exceder esto" o "debes ser al menos aquello" se convierte en una l√≠nea, un plano, o una superficie curva que delimita lo permitido de lo prohibido. Y de la pura abstracci√≥n de infinitas decisiones, emergen **figuras geom√©tricas**.

Nuestro universo de posibilidades se contrae, se define, y lo que queda es un subconjunto contorneado: el lugar donde nuestras decisiones son no solo plausibles, sino **v√°lidas**. Es como si una mano invisible estuviera trazando los contornos de nuestro destino.


```python
# Create the plot
plt.figure(figsize=(8, 8))

# Define the limits for the imshow background
xmin, xmax = -0.5, 6
ymin, ymax = -0.5, 6

# Create a background grid for imshow (single color)
background_color = np.full((10, 10, 3), [1.0, 0.85, 0.7], dtype=float) # Light peach/orange color
plt.imshow(background_color, extent=[xmin, xmax, ymin, ymax], origin='lower', alpha=0.5)

# Plot the axes
plt.axvline(x=0, color='gray', linestyle='-', linewidth=0.8)
plt.axhline(y=0, color='gray', linestyle='-', linewidth=0.8)

# Plot each point and label it
for label, (x1, x2) in points.items():
    plt.plot(x1, x2, 'o', markersize=8, label=label)
    plt.text(x1 + 0.1, x2 + 0.1, f'{x1} manzanas, {x2} pl√°tanos', fontsize=9)

# --- A√±adimos las restricciones y la regi√≥n factible ---

# Definimos el rango para x1 (manzanas) en el que se dibujar√°n las l√≠neas de restricci√≥n
x1_plot_range = np.linspace(0, xmax, 400)

# Restricci√≥n 1: Presupuesto (2*x1 + 1*x2 <= 10) => x2 <= 10 - 2*x1
x2_budget = 10 - 2 * x1_plot_range

# Restricci√≥n 2: Cantidad M√°xima (x1 + x2 <= 6) => x2 <= 6 - x1
x2_max_qty = 6 - x1_plot_range

# Trazamos las l√≠neas de las restricciones
plt.plot(x1_plot_range, x2_budget, label=r'Presupuesto: $2x_1 + x_2 \leq 10$', color='red', linestyle='--')
plt.plot(x1_plot_range, x2_max_qty, label=r'Cantidad M√°x: $x_1 + x_2 \leq 6$', color='green', linestyle='--')

# Calculamos la regi√≥n factible (donde se cumplen todas las restricciones y la no negatividad)
# La regi√≥n factible est√° limitada por x1 >= 0, x2 >= 0, y el m√≠nimo de las l√≠neas de restricci√≥n
x2_min_constraints = np.minimum(x2_budget, x2_max_qty)
x2_feasible_region = np.maximum(0, x2_min_constraints) # Aseguramos que x2 >= 0

# Creamos una m√°scara para que x1 tambi√©n sea >= 0
mask_x1_ge_0 = x1_plot_range >= 0

# Rellenamos la regi√≥n factible
plt.fill_between(x1_plot_range, 0, x2_feasible_region, where=mask_x1_ge_0, color='lightblue', alpha=0.7, label='Regi√≥n Factible')

# --- Fin de las restricciones ---

# Set plot limits and labels
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.xlabel(r'$x_1$ (Kilos de Manzanas)')
plt.ylabel(r'$x_2$ (Kilos de Pl√°tanos)')
plt.title('Espacio de Decisiones y Regi√≥n Factible (Problema de la Fruta)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Leyenda', loc='upper right')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# La Funci√≥n Objetivo
> $ùëì(ùö°‚Éó)$   

> Tu Br√∫jula en el Mapa de Decisiones

La **Funci√≥n Objetivo**. Es la diana de cualquier problema de optimizaci√≥n. Si el **Espacio de Decisi√≥n** es todo el mapa, y la **Regi√≥n Factible** es los mares por donde tu barco puede navegar *realmente*, la Funci√≥n Objetivo es esa **br√∫jula m√°gica** que te dice: "¬°Por aqu√≠ est√° el tesoro!" o "¬°Por aqu√≠ te alejas del peligro!".

Imagina que eres un explorador. No solo quieres estar en un lugar posible, quieres el *mejor* lugar. ¬øQu√© significa "mejor"? Eso lo define tu Funci√≥n Objetivo.

*   ¬øBuscas la cima de la monta√±a m√°s alta (maximizar beneficio, satisfacci√≥n)?
*   ¬øO el punto m√°s bajo en el valle (minimizar coste, riesgo, error)?

En esencia, la Funci√≥n Objetivo ($f(\mathbf{x})$) es una **calculadora de valor**. Toma cualquiera de tus decisiones posibles (un punto $\mathbf{x}$ en tu espacio de decisi√≥n, como `x1` y `x2` en nuestro problema de la fruta) y te devuelve un n√∫mero: la "puntuaci√≥n" de esa decisi√≥n. Es un mapa de calor que se superpone a tu regi√≥n factible, donde cada color (o cada valor de $Z$) representa qu√© tan buena es esa combinaci√≥n particular de $\mathbf{x}$s.

Matem√°ticamente, lo expresamos como:

$$ Z = f(\mathbf{x}) $$

donde $\mathbf{x}$ es el vector de tus variables de decisi√≥n. Queremos encontrar la $\mathbf{x}$ dentro de nuestra **Regi√≥n Factible** que hace que este $Z$ sea el m√°s grande o el m√°s peque√±o posible.



## Conjunto Factible:
> Los mares donde viajas.

Cuando tienes tienes un mont√≥n de reglas en este juego de la optimizaci√≥n, cada regla es una **restricci√≥n**. ¬øRecuerdas ese "universo de posibilidades" que era nuestro espacio de decisi√≥n? Pues bien, cada restricci√≥n es como una pared, o un corte, un muro que te impide ir m√°s all√°.

Si un mapa te indica el territorio, tu barco te restringe a los mares.

### Intersecciones:


Si tengo una regla que dice "no puedes gastar m√°s de 10 euros", eso corta mi universo frutal, ¬øverdad? Y si a√±ado otra regla: "tampoco puedes comprar m√°s de 5 kilos", eso hace *otro corte*. Donde esos cortes se encuentran, donde se **intersectan**, ah√≠ est√° la clave. Cada restricci√≥n lineal corta un pedazo de nuestro espacio, dejando solo el lado "permitido". Es como si cada "no puedes hacer esto" y "tienes que hacer aquello" fuera una tijera que recorta el lienzo.

###  Poliedros:
> ¬°La forma que queda!

Cuando tienes muchas de estas **intersecciones** de restricciones lineales, ¬øqu√© forma obtienes? Pues, si te fijas, obtienes una figura con "caras planas". En matem√°ticas, a estas figuras de muchas caras les llamamos **poliedros**.

Piensa en un cubo, o en una pir√°mide, o en cualquier forma hecha con l√≠neas rectas y superficies planas. Ese es el "espacio permitido" que nos dejan las reglas.

### Regiones: Nuestro patio de juegos

As√≠ que, despu√©s de todos esos recortes (las intersecciones), la forma que queda (el poliedro) es nuestra **regi√≥n factible**. Este es el *verdadero* patio de juegos, el lugar donde todas las reglas se cumplen a la vez. Cada punto dentro de esta regi√≥n es una soluci√≥n v√°lida. Si tu problema es encontrar el mejor desayuno, tu regi√≥n factible es el conjunto de todas las combinaciones de fruta que puedes comprar y que cumplen con tu presupuesto y tu capacidad para llevarlas.

**En resumen**: Empiezas con todo un universo de opciones, las restricciones van haciendo **intersecciones**, estas intersecciones **moldean un poliedro**, y ese poliedro es nuestra **regi√≥n factible**. Es el espacio donde buscamos nuestra soluci√≥n √≥ptima, ¬°ni m√°s ni menos!

# TEMA 2
# Estructura Matem√°tica de un Problema

### *De la vida real al lenguaje formal.*

## 2.1 El Molde Universal de un Problema:  

Imagina que tienes un problema en la vida real. La vida real suele, de echo, traer muchos problemas. Podr√≠a ser cualquier cosa: c√≥mo llegar m√°s r√°pido al trabajo, c√≥mo hacer la mejor tarta con lo que tienes en la nevera, o c√≥mo mandar una nave espacial a Marte con el menor combustible.

Todos estos problemas, por muy diferentes que parezcan, tienen una "receta" b√°sica, un "molde universal" en optimizaci√≥n. Si entiendes este molde, entiendes c√≥mo atacar casi cualquier problema.

### *1¬∫ ¬øQu√© puedes cambiar?*   Las **Variables**

Piensa en tu coche. Puedes pisar m√°s o menos el acelerador, girar el volante... esas son las cosas que controlas. En una tarta, la cantidad de az√∫car, de harina. Esas son tus **variables**. Son las decisiones que tienes que tomar, las "palancas" que puedes mover. En matem√°ticas, las llamamos $x_1, x_2, \dots, x_n$. Simples, ¬øverdad?

### *¬øHasta d√≥nde puedes cambiarlo?* ‚Äî Los **Dominios**

No puedes pisar el acelerador infinitamente, hay un l√≠mite. Tampoco puedes poner medio kilo de az√∫car si solo tienes cien gramos. El **dominio** de una variable es simplemente el rango de valores que puede tomar. ¬øEs un n√∫mero entero? ¬øPuede ser cualquier n√∫mero real? ¬øSolo positivo? Es la "caja" donde vive cada una de tus decisiones.

### *¬øQu√© reglas hay que seguir?* ‚Äî Las **Restricciones**

En la vida, hay reglas. No puedes ir a 200 km/h en una zona escolar. No puedes usar un ingrediente que no tienes. Las **restricciones** son esas reglas. Son ecuaciones o inecuaciones que te dicen qu√© combinaciones de tus variables son "legales" y cu√°les no. Son las paredes de tu "caja de decisiones".

### *¬øQu√© quieres conseguir?* ‚Äî La **Funci√≥n Objetivo**

¬øQuieres llegar m√°s r√°pido? ¬øQuieres que la tarta est√© deliciosa? ¬øQuieres gastar menos combustible? Eso es tu **funci√≥n objetivo**. Es la "puntuaci√≥n" de tu juego. Toma tus variables (tus decisiones) y te devuelve un n√∫mero: lo bueno (o malo) que es tu resultado. En matem√°ticas, la llamamos $f(x_1, x_2, \dots, x_n)$. Queremos que este n√∫mero sea lo m√°s grande o lo m√°s peque√±o posible.
