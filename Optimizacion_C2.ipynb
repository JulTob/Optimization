## 2.1 Las variables como coordenadas del mundo posible

En el fascinante universo de la optimización, las decisiones que tomamos en el mundo real pueden ser formalizadas y representadas de una manera sorprendentemente elegante: a través de las matemáticas. De hecho, `la matemática nace de las ideas`, y en este contexto, la idea fundamental es que cada elección posible puede ser vista como un punto en un "mundo" abstracto, un espacio matemático.

Imaginemos que cada factor sobre el que podemos decidir es una `variable de decisión`, denotada como $x_i$. Por ejemplo, $x_1$ podría ser la cantidad de un producto a fabricar, $x_2$ la inversión en marketing, y así sucesivamente. Cuando combinamos todas estas variables individuales, obtenemos un `vector de decisión` $\mathbf{x} = (x_1, x_2, \ldots, x_n)$. Este vector $\mathbf{x}$ no es solo una lista de números; es una `coordenada` única que ubica una posible decisión en un `espacio matemático` de $n$ dimensiones. Cada dimensión de este espacio corresponde a una de nuestras variables de decisión.

El `significado de estas coordenadas` es profundo: cada conjunto de valores para $(x_1, x_2, \ldots, x_n)$ representa un estado específico, una configuración única de nuestras decisiones. Las `propiedades` de estas variables son cruciales: pueden ser `continuas` (como el tiempo o la cantidad de un líquido), `discretas` (como el número de empleados), o incluso `binarias` (sí/no, 0/1). Estas propiedades definen la naturaleza del espacio de decisión; un espacio con variables continuas es liso y permite infinitas posibilidades entre dos puntos, mientras que uno con variables discretas o binarias es más granular, con puntos definidos y separados.

De esta manera, el vector $\mathbf{x}$ y el espacio que lo contiene `definen el conjunto de todas las decisiones posibles`. Este `mundo posible` es el lienzo sobre el cual la optimización busca el "mejor" punto, la mejor combinación de decisiones que nos lleve a cumplir nuestros objetivos.

La **función objetivo**, denotada como $f(x)$, es el núcleo de cualquier problema de optimización continua. Representa una medida cuantitativa de la "calidad", "costo", "beneficio", o "rendimiento" que deseamos optimizar. En esencia, $f(x)$ toma como entrada un conjunto de decisiones o parámetros, $x$, y produce un valor numérico que nos indica cuán "buena" es esa configuración particular.

Para ilustrar esto, consideremos algunos **ejemplos intuitivos**:
*   **Minimizar costos de producción:** Una empresa desea encontrar la combinación óptima de recursos ($x$) para producir un bien, donde $f(x)$ sería el costo total de producción. El objetivo es minimizar $f(x)$.
*   **Maximizar ganancias:** En un problema de inversión, $x$ podría representar la asignación de capital a diferentes activos, y $f(x)$ sería la ganancia esperada. Aquí, buscamos maximizar $f(x)$.
*   **Minimizar tiempo de viaje:** Para una ruta de transporte, $x$ podrían ser las decisiones sobre qué carreteras tomar, y $f(x)$ el tiempo total de viaje. El objetivo es minimizar $f(x)$.
*   **Optimizar el diseño de un puente:** $x$ serían las dimensiones y materiales, y $f(x)$ podría ser una métrica que combine resistencia, costo y peso. Podríamos querer minimizar el peso o costo mientras se mantiene una resistencia mínima.

Esta función es fundamental porque **dirige el proceso de optimización**. El algoritmo de optimización tiene como misión principal encontrar el valor (o conjunto de valores) de $x$ que resulta en el valor deseado (máximo o mínimo) de $f(x)$. En otras palabras, la optimización es el proceso de buscar el $x^*$ tal que $f(x^*)$ es el valor óptimo (global o local). Esta búsqueda integra la intuición de "hacer algo lo mejor posible" con una formulación matemática rigurosa que permite a los algoritmos explorar el espacio de soluciones de manera eficiente.

## 2.3 Restricciones: Los Límites de lo Posible

En el mundo real, las decisiones no se toman en el vacío; siempre están sujetas a una serie de limitaciones y reglas. Estas pueden ser tan variadas como un presupuesto fijo para un proyecto, la disponibilidad limitada de materias primas, la capacidad máxima de producción de una fábrica, o incluso regulaciones ambientales. En el contexto de la optimización continua, estas limitaciones del mundo real se traducen directamente en **restricciones matemáticas**.

### Traducción de Limitaciones a Restricciones Matemáticas

Las restricciones matemáticas son ecuaciones o desigualdades que los variables de decisión (representados por el vector `x`) deben satisfacer. Por ejemplo:

*   **Restricciones de Desigualdad:** Representan límites superiores o inferiores. Si tienes un presupuesto de $1000, y `x_1` y `x_2` son los costos de dos ítems, la restricción podría ser `x_1 + x_2 <= 1000`. En una forma general, esto se expresa como `g(x) <= 0` (o `g(x) >= 0`), donde `g(x)` es una función que evalúa alguna característica de las variables de decisión.

*   **Restricciones de Igualdad:** Representan condiciones que deben cumplirse exactamente. Por ejemplo, si una receta requiere que la suma de dos ingredientes sea exactamente 5 unidades, la restricción sería `x_1 + x_2 = 5`. En su forma general, esto se expresa como `h(x) = 0`, donde `h(x)` es una función que debe ser cero para que la condición se cumpla.

Estas restricciones son fundamentales porque dictan qué soluciones son permisibles y cuáles no lo son. Sin ellas, el problema de optimización podría tener una solución trivial o no realista (por ejemplo, producir una cantidad infinita de un producto).

### La Región Factible

Cuando todas las restricciones matemáticas se aplican a las variables de decisión, definen un subconjunto específico dentro del espacio total de posibles soluciones. Este subconjunto se conoce como la **región factible** (o *feasible region*).

*   **Definición:** La región factible es el conjunto de todos los puntos (vectores `x`) que satisfacen simultáneamente todas las restricciones del problema de optimización. Cualquier punto `x` dentro de esta región es una solución 'factible' o 'válida', lo que significa que cumple con todas las limitaciones del mundo real.

*   **Importancia:** La región factible es crucial porque el objetivo de la optimización es encontrar el mejor valor posible para la función objetivo, pero *solo* entre aquellos puntos que residen dentro de esta región. Es decir, la solución óptima debe ser un punto factible. Si un problema no tiene ninguna solución que satisfaga todas las restricciones, entonces la región factible está vacía, y el problema no tiene solución factible.

### Formulación Matemática General de un Problema de Optimización Continua

Un problema de optimización continua busca encontrar los valores de un conjunto de variables de decisión que minimicen (o maximicen) una función objetivo, mientras se satisfacen un conjunto de restricciones. La formulación estándar se expresa de la siguiente manera:

$$\begin{array}{ll}
\text{minimizar} & f(\mathbf{x}) \\
\text{sujeto a} & g_i(\mathbf{x}) \le 0, \quad i = 1, \dots, m \\
& h_j(\mathbf{x}) = 0, \quad j = 1, \dots, p
\end{array}$$

Donde:

*   **$\mathbf{x} \in \mathbb{R}^n$**: Es el vector de variables de decisión. Cada componente $x_k$ representa una cantidad que podemos ajustar para lograr nuestro objetivo. En el contexto de problemas del mundo real, estas variables podrían ser, por ejemplo, las cantidades a producir de diferentes productos, los niveles de inversión en distintos activos, o las dimensiones de un diseño de ingeniería.

*   **$f(\mathbf{x}): \mathbb{R}^n \to \mathbb{R}$**: Es la función objetivo (también conocida como función de costo o función a minimizar). Esta función cuantifica el desempeño o la "calidad" de una solución dada por $\mathbf{x}$. Nuestro objetivo es encontrar un $\mathbf{x}$ que produzca el valor más bajo posible para $f(\mathbf{x})$ (si el objetivo es maximizar, se puede transformar fácilmente a minimizar $-f(\mathbf{x})$). En la práctica, $f(\mathbf{x})$ podría representar el costo total de producción, el tiempo de entrega, el riesgo financiero, o cualquier otra métrica que se busque optimizar.

*   **$g_i(\mathbf{x}) \le 0, \quad i = 1, \dots, m$**: Son las restricciones de desigualdad. Estas definen el espacio de soluciones factibles al establecer límites superiores o inferiores en ciertas combinaciones de las variables de decisión. Cada $g_i(\mathbf{x})$ representa una condición que debe cumplirse. Un ejemplo en el mundo real podría ser una restricción de presupuesto (el costo total no debe exceder un límite), la capacidad de una máquina (la producción no puede exceder la capacidad máxima), o la disponibilidad de materia prima.

*   **$h_j(\mathbf{x}) = 0, \quad j = 1, \dots, p$**: Son las restricciones de igualdad. Estas imponen condiciones exactas que deben cumplir las variables de decisión. Cada $h_j(\mathbf{x})$ debe ser cero en la solución óptima. Estas restricciones a menudo representan balances, como leyes de conservación (e.g., la suma de las partes debe ser igual al total) o requisitos de producción exactos.

### Combinación de la Función Objetivo y las Restricciones

La formulación estándar combina estos elementos de manera que el problema se entiende como la búsqueda de un punto $\mathbf{x}$ que no solo minimiza (o maximiza) $f(\mathbf{x})$, sino que también reside dentro de la región factible definida por el conjunto de restricciones de desigualdad ($g_i(\mathbf{x}) \le 0$) y restricciones de igualdad ($h_j(\mathbf{x}) = 0$). La función objetivo guía la búsqueda del "mejor" punto, mientras que las restricciones aseguran que el punto "mejor" sea también "válido" y "alcanzable" dentro de las limitaciones del sistema que se está modelando. Esta combinación es fundamental para modelar decisiones complejas del mundo real, donde siempre hay un objetivo a alcanzar sujeto a diversas limitaciones y recursos.

Para comenzar a trabajar con optimización continua en Python, es fundamental entender cómo podemos representar los elementos básicos de un problema: las **variables de decisión** y la **función objetivo**.

#### Representando Variables de Decisión (vector x)
En problemas de optimización, nuestras variables de decisión a menudo forman un vector. Python, con la ayuda de la librería `NumPy`, es excelente para manejar este tipo de estructuras. Un vector de variables, como `x`, se puede representar fácilmente utilizando un `numpy.array`. Por ejemplo, si tenemos un problema con tres variables `x1, x2, x3`, podemos definirlas como `x = np.array([x1_valor, x2_valor, x3_valor])`.

#### Definiendo Funciones Objetivo Simples
La función objetivo `f(x)` es lo que queremos minimizar o maximizar. En Python, podemos definir estas funciones usando la sintaxis `def`. Por ahora, nos centraremos en funciones simples para ilustrar el concepto, sin introducir algoritmos complejos de optimización.

Por ejemplo, una función objetivo muy básica podría ser `f(x) = x[0]**2 + x[1]**2`, donde `x[0]` y `x[1]` son los primeros elementos de nuestro vector de variables de decisión `x`.
