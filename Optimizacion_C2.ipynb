###  Qu es un Vector de Variables?  Una lista de decisiones

Imagina que ests en la cocina, intentando hornear el pastel perfecto. No solo ajustas la cantidad de azcar, verdad? Tambin ajustas la harina, los huevos, la temperatura del horno, el tiempo de coccin un montn de cosas! Cada una de estas cosas que puedes ajustar es una "variable".

Ahora, en lugar de pensar en cada una de ellas por separado, qu pasara si las metiramos todas en una misma lista? Esa lista, en matemticas, es lo que llamamos un **vector de variables**.

La naturaleza no solo tiene una cosa que cambia. Tiene muchas cosas que cambian al mismo tiempo. Y cuando queremos hablar de cmo afectan al sistema, no vamos a listarlas una por una. Eso es una prdida de tiempo. Las agrupamos. Las empaquetamos. Las llamamos un vector.

**Formalmente:**

En el mundo de la optimizacin y el clculo, cuando hablamos de un problema con mltiples decisiones interrelacionadas, representamos esas decisiones como un **vector de variables de decisin**. Si tenemos 'n' decisiones que tomar, podemos agruparlas en un solo objeto matemtico:

$$\mathbf{x} = \begin{pmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{pmatrix}$$

Donde:

*   **$\mathbf{x}$** (en negrita) representa el vector completo de variables.
*   **$x_1, x_2, \dots, x_n$** son las **componentes** individuales del vector, y cada una es una variable de decisin diferente (por ejemplo, $x_1$ podra ser la cantidad de azcar, $x_2$ la cantidad de harina, etc.).
*   Este vector \(\mathbf{x}\) pertenece a un espacio de 'n' dimensiones, lo que denotamos como **$\mathbf{x} \in \mathbb{R}^n$**. Esto simplemente significa que cada una de las 'n' variables puede tomar cualquier valor real, .

As que, cuando ves \(\mathbf{x}\) en una ecuacin, no pienses en una sola cosa. Piensa en esa "caja" llena de todas las decisiones que tienes que tomar simultneamente. Es la forma eficiente de la matemtica de decir: "Aqu est el conjunto completo de palancas que puedes mover para influir en tu sistema".

### Hasta dnde puedes cambiarlo?  Los **Dominios**

En el universo no todo vale. No puedes pisar el acelerador de tu coche infinitamente. Hay un lmite universal $(c)$ y legal $(120km/h)$!. No se puede poner medio kilo de azcar en la tarta si solo tienes cien gramos. Hay reglas, lmites fsicos, o simplemente, de disponibilidad.

Esos lmites, esa "regin" dentro de la cual cada una de tus variables puede moverse, es lo que llamamos el **dominio**. Es el conjunto de todos los valores posibles que una variable (o un conjunto de variables) puede tomar en el contexto de tu problema.

**Formalmente:**

Si tenemos una funcin $f(\mathbf{x})$, donde $\mathbf{x} = (x_1, x_2, \dots, x_n)$ es nuestro vector de variables de decisin, el **dominio** de esa funcin (y por extensin, de nuestras variables) es el conjunto de todos los $(\mathbf{x})$ para los cuales la funcin est definida y es significativa para nuestro problema. Lo denotamos como $\mathbb{D}$.

Por ejemplo:

*   Si $x_1$ es la cantidad de un ingrediente, su dominio natural podra ser $x_1 \ge 0$, porque no puedes tener una cantidad negativa.
*   Si $x_2$ es la temperatura del horno, su dominio podra ser $[150, 250]$ grados Celsius, si sabemos que hornear fuera de ese rango daa el pastel o no cocina adecuadamente.

Matemticamente, para una sola variable $x$, el dominio podra expresarse como:

- $x \in [a, b]$ (un intervalo cerrado)
- $x \in (a, b)$ (un intervalo abierto)
- $x \in \mathbb{R}$ (todos los nmeros reales)
- $x \in \mathbb{Z}$ (todos los nmeros enteros)

Para un vector de variables $\mathbf{x} \in \mathbb{R}^n$, el dominio $\mathbb{D}$ es un subconjunto de $\mathbb{R}^n$.

Por ejemplo, si tenemos $x_1$ y $x_2$, y $x_1 \ge 0$ y $x_2 \ge 0$, entonces el dominio es el primer cuadrante del plano cartesiano: $\mathbb{D} = \{(x_1, x_2) \in \mathbb{R}^2 \mid x_1 \ge 0, x_2 \ge 0\}$.

As que, el dominio es simplemente el 'terreno de juego' para cada una de tus decisiones.


### Qu reglas hay que seguir?  Las **Restricciones**

La vida no solo aporta problemas, tambin aporta limitaciones. Cuando queremos viajar tenemos que asegurarnos que todo quepa en una maleta tamao "Ryanair estandar del ms de Agosto", cuando planeamos nuestra ruta tenemos que considerar el dinero que tenemos disponible, y cuantos das de vacaciones nos _concede_ la empresa.Reglas. Esas condiciones que te dicen qu combinaciones de tus variables son posibles y cules no, son las **restricciones**. Son, por as decirlo, las paredes de tu "caja de decisiones", la regin factible. Sin ellas, tus soluciones seran pura fantasa.

**Formalmente:**

En un problema de optimizacin, las restricciones son condiciones que las variables de decisin deben satisfacer. Pueden ser de dos tipos principales: **desigualdades** e **igualdades**.

Consideremos nuestro vector de variables de decisin $\mathbf{x} \in \mathbb{R}^n$. Las restricciones se expresan como funciones de $\mathbf{x}$:

1.  **Restricciones de Desigualdad:** Estas establecen lmites superiores o inferiores. Por ejemplo, la cantidad de un recurso no puede exceder la disponibilidad, o una produccin debe ser al menos un cierto valor. Se expresan comnmente como:

    $$ g_i(\mathbf{x}) \le 0, \quad \text{para } i = 1, \dots, m $$

    Donde cada $g_i: \mathbb{R}^n \to \mathbb{R}$ es una funcin que toma el vector de decisiones $\mathbf{x}$ y devuelve un nico valor real. La condicin $g_i(\mathbf{x}) \le 0$ define la regin permitida. Si tuviramos un lmite superior como $x_1 + x_2 \le 10$, lo reescribimos como $x_1 + x_2 - 10 \le 0$, y as $g_i(\mathbf{x}) = x_1 + x_2 - 10$.

2.  **Restricciones de Igualdad:** Estas imponen condiciones exactas. Por ejemplo, un balance de materiales o una asignacin de recursos que debe ser precisa. Se formulan como:

    $$ h_j(\mathbf{x}) = 0, \quad \text{para } j = 1, \dots, p $$

    Donde cada $h_j: \mathbb{R}^n \to \mathbb{R}$ es otra funcin. Si, por ejemplo, la suma de dos variables debe ser exactamente 5, esto se representa como $x_1 + x_2 = 5$, lo que se transforma en $h_j(\mathbf{x}) = x_1 + x_2 - 5 = 0$.


Estas funciones, $g_i$ y $h_j$, son solo nuestra manera elegante de escribir las reglas del juego. Son simplemente matemticas que definen lo que se puede y no se puede hacer.

### *El terreno de juego real.*  El **Conjunto Factible**

Tenemos nuestros 'diales' (las variables) y sabemos dnde pueden moverse (sus dominios). Tambin tenemos las 'reglas del juego' (las restricciones). Ahora, qu pasa si lo juntamos todo?

Lo que obtenemos es el **conjunto factible**.

Es la combinacin de todas esas decisiones que son a la vez *posibles* (dentro de los dominios) y *permitidas* (satisfaciendo las restricciones).

Es el verdadero 'campo de juego' donde podemos buscar la mejor solucin. Si lo piensas, es la interseccin de todos tus dominios y todas tus restricciones.

**Formalmente:**

El conjunto factible, denotado como $\mathbb{F}$, es el subconjunto de $\mathbb{R}^n$ que contiene todos los vectores de decisin $\mathbf{x}$ que satisfacen simultneamente todas las restricciones de desigualdad y de igualdad.

Matemticamente, se define como:

$$\mathbb{F} = \{\mathbf{x} \in \mathbb{R}^n \mid g_i(\mathbf{x}) \le 0, \text{ para } i=1,\dots,m \quad \text{y} \quad h_j(\mathbf{x}) = 0, \text{ para } j=1,\dots,p \}$$

Donde:
*   $\mathbf{x}$ es el vector de variables de decisin.
*   $g_i(\mathbf{x}) \le 0$ son las restricciones de desigualdad.
*   $h_j(\mathbf{x}) = 0$ son las restricciones de igualdad.

En resumen, el conjunto factible es donde realmente puedes hacer negocios. Cualquier otra cosa es ilegal. Aqu es donde de verdad empieza la bsqueda del tesoro!.

**Formalmente:**

La funcin objetivo, a menudo denotada como $f(\mathbf{x})$, es una funcin que mapea el vector de variables de decisin a un escalar real, es decir, $f: \mathbb{R}^n \to \mathbb{R}$. Nuestro propsito es encontrar un vector de decisin ptimo $\mathbf{x}^*$ dentro del conjunto factible $\mathbb{F}$ que minimice o maximice esta funcin.

    Aqu buscamos el valor ms pequeo de $f(\mathbf{x})$. Por ejemplo, minimizar costos o tiempo.

    $$\max_{\mathbf{x} \in \mathbb{F}}:  f(\mathbf{x})$$
    Aqu buscamos el valor ms grande de $f(\mathbf{x})$. Por ejemplo, maximizar ganancias o eficiencia.

La funcin objetivo es el corazn del problema. Es lo que nos dice si estamos ganando o perdiendo en este  juego de la optimizacin.

### *Armando el Rompecabezas:* Qu es un Problema de Optimizacin

Hemos estado desgranando las piezas: los botones que podemos mover (variables), dnde podemos moverlas (dominio), las reglas que nos detienen (restricciones), y lo que queremos lograr (funcin objetivo).  

Pues bien, un **problema de optimizacin** no es ms que el acto de buscar la *mejor* jugada, la *mejor* configuracin de botones, pero siempre, siempre, dentro de las reglas que nos impone el juego.

**Formalmente:**

Un problema de optimizacin continua es la tarea de encontrar un vector de variables de decisin $\mathbf{x}$ dentro de un conjunto factible $\mathbb{F}$ que minimice (o maximice) una funcin objetivo $f(\mathbf{x})$. La formulacin estndar se expresa de la siguiente manera:

$$\begin{array}{ll}
\text{minimizar: } & &f(\mathbf{x}) \\
\text{sujeto a:} & &g_i(\mathbf{x}) \le 0, \quad i = 1, \dots, m \\
&& h_j(\mathbf{x}) = 0, \quad j = 1, \dots, p \\
&& \mathbf{x} \in \mathbb{D}
\end{array}$$

Donde:

*   **$\mathbf{x} \in \mathbb{R}^n$**: Es el vector de variables de decisin, nuestros 'botones'.
*   **$f(\mathbf{x}): \mathbb{R}^n \to \mathbb{R}$**: Es la funcin objetivo, la 'puntuacin' que buscamos optimizar (minimizar o maximizar).
*   **$g_i(\mathbf{x}) \le 0, \quad i = 1, \dots, m$**: Son las restricciones de desigualdad, las 'paredes' que nos impiden ir a ciertos lugares.
*   **$h_j(\mathbf{x}) = 0, \quad j = 1, \dots, p$**: Son las restricciones de igualdad, las 'lneas exactas' que debemos seguir.
*   **$\mathbf{x} \in \mathbb{D}$**: Implcitamente, nuestras variables deben estar dentro de su dominio natural $\mathbb{D}$. A menudo, las restricciones $g_i$ ya capturan estos lmites, pero es importante recordar que cada variable tiene su propio 'espacio vital'.

Un problema de optimizacin es poner a prueba los lmites de lo posible para alcanzar lo ptimo. Es una bsqueda elegante de la mejor manera de operar, bajo las leyes del universo... o, al menos, las del juego.

```python
import numpy as np # Import numpy to use np.inf
import sympy as sym # Import sympy to use symbols

class Decision:
    def __init__(self,
                 symbol,
                 value= None,
                 lower=0,
                 upper=np.inf,
                 numeric = "Real" ):
        self.name = symbol       # "x", "manzanas"
        self.symbol = sym.Symbol(symbol)
        self.lower = lower       # 0
        self.upper = upper       # 6
        self.value = value
        self.numeric = numeric

    def __call__(self):
        return self.symbol

    def __contains__(self, x):
        return self.lower <= x <= self.upper

    def in_domain(self, x):
        return x in self

    def __str__(self):
        if self.value is not None:
            return f"{self.name}  [{self.lower}, {self.upper}] := {self.value}"
        else:
            return f"{self.name}  [{self.lower}, {self.upper}]"

    def set_upper(self, value):
        self <= value

    def set_lower(self, value):
        self >= value

    def __le__(self, value): # self <= value -- Limits upper
        if value < self.lower:
            raise ValueError(f"{self.name} is unfeasible")

        self.upper = min(value, self.upper)
        return self

    def __ge__(self, value): # self >= value -- Limits lower
        if value > self.upper:
                raise ValueError(f"{self.name} is unfeasible")
        self.lower = max(value, self.lower)
        return self

    def domain(self):
        return (self.lower, self.upper)

    def __abs__(self):
        return abs(self.upper - self.lower)

    def __len__(self):
        return self.upper - self.lower

    def Integer(self):
        self.numeric = "Integer"
        if self.value is not None:
            self.value = int(self.value)
        return self # Return self instead of self.value

    def Real(self)	:
        self.numeric = "Real"
        if self.value is not None:
            self.value = float(self.value)
        return self # Return self instead of self.value

    def Binary(self):
        self.numeric = "Integer"
        self.lower = 0
        self.upper = 1
        return self


class Optimization:
    def __init__(self, variables, function=None, kind = "min"):
        self.variables = variables      # lista de Variable
        self.objective = None           # funcin sympy
        self.kind = None                # "min" o "max"
        self.restrictions = []          # g(x) boolean functions

    def symbols(self):
        return [v.symbol for v in self.variables]

    def set_maximize(self):
        self.kind = "max"

    def set_minimize(self):
        self.kind = "min"

    def __call__(self, f):
        self.objective = f
        return self


    def set_objetivo(self, function, kind="min"):
        self.objective = function
        self.kind = kind

    def __pos__(self):
        self.set_maximize()
        return self

    def __neg__(self):
        self.set_minimize()
        return self

    def add_restriction(self, restriction):
        self.restrictions.append(restriction)

    def __getitem__(self, restriction):
        self.add_restriction(restriction)

    def __str__(self):
        vars = ", ".join([str(x) for x in self.variables])
        X = ", ".join([str(x.name) for x in self.variables])
        rest = "\n\t".join([str(x) for x in self.restrictions])
        return f"""f({X}) = {self.objective}
        {self.kind}imization: {vars}
        s.t.
        {rest}"""


    def plot(self,
             resolution=200,
             palette=None,
             numeric = "Real"):
        import numpy as np
        import matplotlib.pyplot as plt
        import sympy as sym
        from matplotlib.colors import rgb_to_hsv, hsv_to_rgb
        dot_alpha = 1
        if palette is None:
            palette = "gray"
            dot_alpha = 0.7


        # === 1. Variables: lmites de dominio ===
        xs = self.symbols()
        d1, d2 = self.variables  # asumimos 2D por ahora
        xmin, xmax = 0, d1.upper
        ymin, ymax = 0, d2.upper

        # === 2. Crear malla cartesiana ===
        if d1.numeric == "Real":
            x_vals = np.linspace(xmin, xmax, resolution)
        elif d1.numeric == "Integer":
            x_vals = np.arange(int(xmin), int(xmax) + 1)
        else:
            raise ValueError("Numeric type not supported")

        if d2.numeric == "Real":
            y_vals = np.linspace(ymin, ymax, resolution)
        elif d2.numeric == "Integer":
            y_vals = np.arange(int(ymin), int(ymax) + 1)
        else:
            raise ValueError("Numeric type not supported")

        X, Y = np.meshgrid(x_vals, y_vals)

        # === 3. Convertir funcin objetivo a NumPy ===
        if self.objective is None:
            raise ValueError("No objective function defined.")

        f_num = sym.lambdify(xs, self.objective, "numpy")
        Z = f_num(X, Y)

        # === 4. Evaluar restricciones en la malla ===
        masks = []
        for r in self.restrictions:
            g_num = sym.lambdify(xs, r.lhs - r.rhs, "numpy")  # g(x)<=0  g(x)=lhs-rhs
            G = g_num(X, Y)
            masks.append(G <= 0)

        # === 4.5. Mscara del conjunto factible (todas las restricciones a la vez) ===
        feasible_mask = np.ones_like(X, dtype=bool)
        for mask in masks:
            feasible_mask &= mask

        # === 5. Fondo de restricciones con mezcla en HSV ===
        fig, ax = plt.subplots(figsize=(8, 8))

        # Colores base en RGB (01)
        colors_rgb = np.array([
            [0.0, 1.0, 0.0],  # Green
            [1.0, 0.0, 0.0],  # Red
            [0.0, 0.0, 1.0],  # Blue
            [1.0, 1.0, 0.0],  # Yellow
            [1.0, 0.0, 1.0],  # Magenta
            [0.0, 1.0, 1.0],  # Cyan
        ])

        colors_hsv = rgb_to_hsv(colors_rgb)  # mismo shape (n_colors, 3)

        # Preparamos acumuladores del mismo tamao que X/Y
        shape = X.shape
        sum_x = np.zeros(shape, dtype=float)   # para cos(hue)
        sum_y = np.zeros(shape, dtype=float)   # para sin(hue)
        count = np.zeros(shape, dtype=float)   # n de restricciones activas

        for i, mask in enumerate(masks):
            h, s, v = colors_hsv[i % len(colors_hsv)]
            angle = 2 * np.pi * h

            cos_h = np.cos(angle)
            sin_h = np.sin(angle)

            # Acumulamos solo donde la restriccin es vlida
            sum_x[mask] += cos_h
            sum_y[mask] += sin_h
            count[mask] += 1.0

        # Arrays HSV del fondo
        h_bg = np.zeros(shape, dtype=float)
        s_bg = np.zeros(shape, dtype=float)
        v_bg = np.zeros(shape, dtype=float)

        active = count > 0

        # Tono = media circular
        angles = np.arctan2(sum_y[active], sum_x[active])
        h_bg[active] = (angles % (2 * np.pi)) / (2 * np.pi)

        # Saturacin mxima siempre que haya restricciones
        s_bg[active] = 1.0

        # Valor: controlamos brillo para no matar la escala de grises
        # por ejemplo: base 0.4 + 0.15 por restriccin, cap a 0.8
        v_bg[active] = np.clip(0.4 + 0.15 * (count[active] - 1), 0.3, 0.8)

        #  hacer el conjunto factible completamente blanco
        # blanco en HSV = cualquier hue, s = 0, v = 1
        s_bg[feasible_mask] = 0.0
        v_bg[feasible_mask] = 1.0

        # Fondo en RGB
        hsv_bg = np.stack([h_bg, s_bg, v_bg], axis=-1)
        rgb_bg = hsv_to_rgb(hsv_bg)

        ax.imshow(
            rgb_bg,
            extent=[xmin, xmax, ymin, ymax],
            origin="lower"
        )

        # === 6. Computar cantidad de restricciones satisfechas ===
        satisfied_count = sum(mask for mask in masks)
        violated = len(masks) - satisfied_count

        # radio de los puntos
        n_points = resolution**2
        base_r = 6e4 / n_points
        R = base_r * satisfied_count / (1 + violated)

        ax.scatter(X.flatten(), Y.flatten(),
                s=R.flatten(),
                c=Z.flatten(),
                cmap=palette,
                edgecolors="none",
                alpha=dot_alpha)

        # === 7. Marcar ptimo ===

        Z_feasible = np.where(feasible_mask, Z, np.nan)
        if self.kind == "max": opt = np.nanargmax(Z_feasible)
        else: opt = np.nanargmin(Z_feasible)
        x_opt = X.flatten()[opt]
        y_opt = Y.flatten()[opt]
        f_opt = Z.flatten()[opt]

        ax.scatter([x_opt], [y_opt],
                color="gold",
                edgecolor="black",
                s=220, zorder=5)

        opt_mask = (Z_feasible == f_opt)
        x_opt_all = X[opt_mask]
        y_opt_all = Y[opt_mask]

        # === 8. Decoracin ===
        ax.set_title(f"{self.kind}:  f({x_opt:.2f}, {y_opt:.2f}) = {f_opt:.2f}")
        ax.set_xlabel(d1.name)
        ax.set_ylabel(d2.name)
        ax.set_xlim(xmin, xmax)
        ax.set_ylim(ymin, ymax)
        ax.grid(True, alpha=0.3)
        ax.legend()
        ax.scatter(
            x_opt_all, y_opt_all,
            color="gold",
            edgecolor="goldenrod",
            s=40,
            zorder=5
            )


        plt.show()


    def field(self,
              resolution=200,
              palette=None,
              numeric="Real"):
        import numpy as np
        import matplotlib.pyplot as plt
        import sympy as sym
        from matplotlib.colors import rgb_to_hsv, hsv_to_rgb

        dot_alpha = 1
        if palette is None:
            palette = "gray"
            dot_alpha = 0.7

        # === 1. Variables: lmites de dominio ===
        xs = self.symbols()
        d1, d2 = self.variables  # asumimos 2D por ahora
        xmin, xmax = 0, d1.upper
        ymin, ymax = 0, d2.upper

        # === 2. Crear malla cartesiana ===
        if d1.numeric == "Real":
            x_vals = np.linspace(xmin, xmax, resolution)
        elif d1.numeric == "Integer":
            x_vals = np.arange(int(xmin), int(xmax) + 1)
        else:
            raise ValueError("Numeric type not supported")

        if d2.numeric == "Real":
            y_vals = np.linspace(ymin, ymax, resolution)
        elif d2.numeric == "Integer":
            y_vals = np.arange(int(ymin), int(xmax) + 1)
        else:
            raise ValueError("Numeric type not supported")

        X, Y = np.meshgrid(x_vals, y_vals)

        # === 3. Convertir funcin objetivo a NumPy ===
        if self.objective is None:
            raise ValueError("No objective function defined.")

        f_num = sym.lambdify(xs, self.objective, "numpy")
        Z = f_num(X, Y)

        # === 4. Evaluar restricciones en la malla ===
        masks = []
        for r in self.restrictions:
            g_num = sym.lambdify(xs, r.lhs - r.rhs, "numpy")  # g(x)<=0  g(x)=lhs-rhs
            G = g_num(X, Y)
            masks.append(G <= 0)

        # === 4.5. Mscara del conjunto factible (todas las restricciones a la vez) ===
        feasible_mask = np.ones_like(X, dtype=bool)
        for mask in masks:
            feasible_mask &= mask

        # === 5. Fondo de restricciones con mezcla en HSV ===
        fig, ax = plt.subplots(figsize=(8, 8))

        colors_rgb = np.array([
            [0.0, 1.0, 0.0],  # Green
            [1.0, 0.0, 0.0],  # Red
            [0.0, 0.0, 1.0],  # Blue
            [1.0, 1.0, 0.0],  # Yellow
            [1.0, 0.0, 1.0],  # Magenta
            [0.0, 1.0, 1.0],  # Cyan
        ])
        colors_hsv = rgb_to_hsv(colors_rgb)

        shape = X.shape
        sum_x = np.zeros(shape, dtype=float)
        sum_y = np.zeros(shape, dtype=float)
        count = np.zeros(shape, dtype=float)

        for i, mask in enumerate(masks):
            h, s, v = colors_hsv[i % len(colors_hsv)]
            angle = 2 * np.pi * h
            cos_h = np.cos(angle)
            sin_h = np.sin(angle)
            sum_x[mask] += cos_h
            sum_y[mask] += sin_h
            count[mask] += 1.0

        h_bg = np.zeros(shape, dtype=float)
        s_bg = np.zeros(shape, dtype=float)
        v_bg = np.zeros(shape, dtype=float)

        active = count > 0
        angles = np.arctan2(sum_y[active], sum_x[active])
        h_bg[active] = (angles % (2 * np.pi)) / (2 * np.pi)
        s_bg[active] = 1.0
        v_bg[active] = np.clip(0.4 + 0.15 * (count[active] - 1), 0.3, 0.8)

        # zona factible en blanco
        s_bg[feasible_mask] = 0.0
        v_bg[feasible_mask] = 1.0

        hsv_bg = np.stack([h_bg, s_bg, v_bg], axis=-1)
        rgb_bg = hsv_to_rgb(hsv_bg)

        ax.imshow(
            rgb_bg,
            extent=[xmin, xmax, ymin, ymax],
            origin="lower"
        )

        # === 6. Campo de gradiente ===
        n_constraints = len(masks)
        n_points = X.size

        # cuntas restricciones se cumplen en cada punto
        if n_constraints > 0:
            satisfied_count = np.zeros_like(X, dtype=float)
            for mask in masks:
                satisfied_count += mask
            ratio = satisfied_count / n_constraints  # entre 0 y 1
        else:
            ratio = np.ones_like(X, dtype=float)

        # factor de tamao: 0.3 fuera, 1.0 dentro
        size_factor = 0.3 + 0.7 * ratio

        # Gradiente numrico
        grad = [sym.diff(self.objective, var) for var in xs]
        gx_num = sym.lambdify(xs, grad[0], "numpy")
        gy_num = sym.lambdify(xs, grad[1], "numpy")

        GX_scalar = gx_num(X, Y)
        GY_scalar = gy_num(X, Y)

        # Ensure GX and GY are arrays of the same shape as X and Y
        if np.isscalar(GX_scalar):
            GX = np.full_like(X, GX_scalar)
        else:
            GX = GX_scalar

        if np.isscalar(GY_scalar):
            GY = np.full_like(Y, GY_scalar)
        else:
            GY = GY_scalar

        # Normalizamos gradiente; longitud depende de size_factor, no de |f|
        norm = np.hypot(GX, GY)
        norm[norm == 0] = 1.0

        # base de longitud ligada al tamao del dominio
        base_len = 0.03 * max(xmax - xmin, ymax - ymin)

        U = GX / norm * base_len * size_factor
        V = GY / norm * base_len * size_factor

        # Submuestreo para no llenarlo todo de flechas
        step = max(1, resolution // 25)

        ax.quiver(
            X[::step, ::step],
            Y[::step, ::step],
            U[::step, ::step],
            V[::step, ::step],
            Z[::step, ::step],       # color segn f(x,y)
            cmap=palette,
            alpha=dot_alpha,
            angles="xy",
            scale_units="xy",
            scale=1.0,
            width=0.004
        )

        # === 7. Marcar ptimo ===
        Z_feasible = np.where(feasible_mask, Z, np.nan)
        if self.kind == "max":
            opt = np.nanargmax(Z_feasible)
        else:
            opt = np.nanargmin(Z_feasible)

        x_opt = X.flatten()[opt]
        y_opt = Y.flatten()[opt]
        f_opt = Z.flatten()[opt]

        ax.scatter(
            [x_opt], [y_opt],
            color="gold",
            edgecolor="black",
            s=220,
            zorder=5
        )

        opt_mask = (Z_feasible == f_opt)
        x_opt_all = X[opt_mask]
        y_opt_all = Y[opt_mask]

        ax.set_title(f"{self.kind}:  f({x_opt:.2f}, {y_opt:.2f}) = {f_opt:.2f}")
        ax.set_xlabel(d1.name)
        ax.set_ylabel(d2.name)
        ax.set_xlim(xmin, xmax)
        ax.set_ylim(ymin, ymax)
        ax.grid(True, alpha=0.3)

        # Si no tienes artistas con label, mejor quitar esto para evitar warning:
        # ax.legend()

        ax.scatter(
            x_opt_all, y_opt_all,
            color="gold",
            edgecolor="goldenrod",
            s=40,
            zorder=5
        )

        plt.show()
```

```python
def classify_point(H, point):
    H_eval = H.subs(point)
    eigenvals = H_eval.eigenvals()

    eigs = list(eigenvals.keys())

    if all(ev > 0 for ev in eigs):
        return "min"
    if all(ev < 0 for ev in eigs):
        return "max"
    if any(ev < 0 for ev in eigs) and any(ev > 0 for ev in eigs):
        return "saddle"
    return "degenerate"
```

```python
import numpy as np
import sympy as sym

def feasible(P, point):
    # Check all restrictions
    for r in P.restrictions:
        try:
            # Substitute point into LHS and RHS of the relational expression
            # This might result in an expression if not all symbols are in 'point'
            lhs_substituted = r.lhs.subs(point)
            rhs_substituted = r.rhs.subs(point)

            # Check if substitution resulted in non-numeric SymPy expressions
            # If so, this 'point' cannot fully evaluate this constraint, so it's not well-defined for this check
            if not (lhs_substituted.is_Number and rhs_substituted.is_Number):
                # If either side is still symbolic, it means the point doesn't fully define the expression
                # For our context (checking feasibility at a specific point), this usually means it's not fully defined
                return False

            lhs_val = float(lhs_substituted)
            rhs_val = float(rhs_substituted)

            # Check violation based on the type of SymPy relational operator
            if isinstance(r, sym.Le): # Constraint is of type A <= B
                if lhs_val > rhs_val: # Violated if A > B
                    return False
            elif isinstance(r, sym.Ge): # Constraint is of type A >= B
                if lhs_val < rhs_val: # Violated if A < B
                    return False
            elif isinstance(r, sym.Eq): # Constraint is of type A == B
                # For equality, use a numerical tolerance (np.isclose)
                if not np.isclose(lhs_val, rhs_val):
                    return False
            else:
                # Should not happen if only Le, Ge, Eq are used, but good for robustness
                raise NotImplementedError(f"Unsupported SymPy Relational type: {type(r)}")

        except (TypeError, AttributeError, ValueError) as e:
            # Catch errors during substitution or float conversion for debugging
            print(f"Warning: Could not evaluate restriction {r} at point {point}. Error: {e}")
            return False

    # Check domain bounds from Decision objects
    for v_dec in P.variables:
        # Ensure the current variable is in the point dictionary
        if v_dec.symbol not in point:
            # If a variable is missing from the point, the point is incomplete for feasibility check.
            return False

        val = point[v_dec.symbol]
        # Convert SymPy numeric types to Python float for comparison with v.lower/upper
        # This handles potential SymPy Integers/Floats from substitution
        if not float(v_dec.lower) <= float(val) <= float(v_dec.upper):
            return False

    return True
```

```python
import itertools
import sympy as sym

def enumerate_vertices(P):
    xs = P.symbols()
    n = len(xs)

    equations = [r.lhs - r.rhs for r in P.restrictions]

    vertices = []

    # take every combination of n equations
    for combo in itertools.combinations(equations, n):
        sol = sym.solve(combo, xs, dict=True)
        for s in sol:
            # keep only numerical complete solutions
            if all(v in s and s[v].is_real for v in xs):
                if feasible(P, s):
                    vertices.append(s)

    return vertices
```

```python
def analytical_optima(OP, steps=True):
    import sympy as sym
    from itertools import combinations
    P = OP
    xs = P.symbols()
    f = P.objective

    # Gradiente
    grad = [sym.diff(f, x) for x in xs]

    # Hessiano
    H = sym.hessian(f, xs)

    if steps:
        print("\nGradiente & Hessiana:")
        print(grad)
        print(H)

    # ====== PUNTOS CRTICOS ======
    critical_points = sym.solve(grad, xs, dict=True)

    if steps:
        print("\nPuntos crticos:")
        print(critical_points)
        print(classify_point(H, critical_points))

    candidates = []

    # Evaluar puntos crticos
    for cp in critical_points:
        if feasible(P, cp):
            fval = float(f.subs(cp))
            kind = classify_point(H, cp)
            candidates.append((cp, fval, kind, "critical"))
            if steps:
                print(f"\nCrtico factible: {cp}, f={fval}, tipo={kind}")
        else:
            if steps:
                print(f"\nCrtico NO factible: {cp}")

    vertices = enumerate_vertices(P)

    for v in vertices:
        if feasible(P,v):
            fval = float(f.subs(v))
            candidates.append((v, fval, "vertex", "vertex"))
            if steps:
                print(f"Vrtice factible: {v}, f={fval}, tipo=vertex")

    # ====== SELECT OPTIMUM ======
    if not candidates:
        print("No feasible points found.")
        return

    # candidates = [(point, fval, kind, source), ...]
    if P.kind == "max":
        p_opt, f_opt, kind_opt, src_opt = max(candidates, key=lambda t: t[1])
    else:
        p_opt, f_opt, kind_opt, src_opt = min(candidates, key=lambda t: t[1])

    print(f"\nptimo {P.kind}: f={f_opt} en {p_opt}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import ipywidgets as widgets
from ipywidgets import interact # Import interact function
import sympy as sym # Ensure sympy is imported for feasible function


def sunplot(P, slider_max=10):
    sliders = set_sliders(P, max=slider_max)
    xs = P.symbols()
    labels = [str(v.symbol) for v in P.variables]
    n = len(labels)

    # Radar geometry
    angles = np.linspace(0, 2*np.pi, n, endpoint=False).tolist()

    def plot_frame(**kwargs):
        # substitute chosen values
        vals = np.array([kwargs[str(v.symbol)] for v in P.variables])

        # Create a point dictionary for the feasible check
        current_point = {sym.Symbol(k): v for k, v in kwargs.items()}

        # Check feasibility of the current point
        is_feasible = feasible(P, current_point) # Assuming 'feasible' function is available in scope
        polygon_color = 'lightgreen' if is_feasible else 'tomato'
        polygon_alpha = 0.25 if is_feasible else 0.4

        # evaluate objective
        f_val = float(P.objective.subs({sym: kwargs[str(sym)] for sym in xs}))

        # === Radar Plot ===
        fig = plt.figure(figsize=(6,6))
        ax = plt.subplot(111, polar=True)

        # Dynamically determine the overall maximum slider value for radar extent
        overall_max = 0
        for slider_key, slider_widget in sliders.items():
            if hasattr(slider_widget, 'max'): # Check if it's a widget with a 'max' attribute
                overall_max = max(overall_max, slider_widget.max)

        # Set radial limits based on overall_max
        ax.set_rlim(0, overall_max)

        # close polygon
        vals_closed = np.concatenate((vals, [vals[0]]))
        angles_closed = angles + [angles[0]]

        # draw radar outline
        ax.plot(angles_closed, vals_closed, linewidth=2, color=polygon_color)
        ax.fill(angles_closed, vals_closed, alpha=polygon_alpha, color=polygon_color)

        # label axes
        ax.set_xticks(angles)
        ax.set_xticklabels(labels)


        # === Sun in center ===
        radius = np.log(1+abs(f_val))
        color = plt.cm.inferno(0.5 + np.tanh(f_val/(overall_max+1)**n))


        # Growth Vectors
        grad = [sym.diff(P.objective, v.symbol) for v in P.variables]
        grad_val = np.array([float(g.subs(current_point)) for g in grad])
        for k, gk in enumerate(grad_val):
            print(gk)
            angle = angles[k]
            color_arr = plt.cm.seismic(0.5 - np.tanh(gk/overall_max))
            ax.annotate(
                "",
                xy=(angle, gk),
                xytext=(angle, 0),
                arrowprops=dict(arrowstyle="->", color=color_arr, linewidth=2)
                )

        sun = Circle((0,0), radius=radius, transform=ax.transData._b, color=color, alpha=0.95)
        ax.add_artist(sun)

        ax.set_title(f"f(x) = {f_val:.3f} {'(Infeasible)' if not is_feasible else ''}", fontsize=14)
        plt.show()

    interact(plot_frame, **sliders)
```

```python
import ipywidgets as widgets

def set_sliders(P, max=10.0, step=0.1, initial=0.0):
    """
    Build FloatSliders for all decision variables in an Optimization problem.

    Parameters
    ----------
    P : Optimization - Your optimization problem
    max : float - Upper bound for all sliders
    step : float - Slider resolution
    initial : float - Initial value for all sliders (default 0)

    Returns
    -------
    dict[str, widgets.FloatSlider]
        Ready to be passed into sunplot or interact
    """
    sliders = {}

    for v in P.variables:
        if v.numeric == "Integer": step = 1
        sliders[str(v.symbol)] = widgets.FloatSlider(
            value=initial,
            min=0.0,
            max=max,
            step=step,
            description=str(v.symbol),
            continuous_update=True,
            readout_format=".2f"
        )

    return sliders
```

```python
import itertools
import sympy as sym
import numpy as np # Import numpy for np.inf

def integer_optima(OP):
    variables = OP.variables

    # 1. construir dominios
    domains = []
    for x in variables:
        low = x.lower
        high = x.upper

        # Enforce finite bounds for integer variables
        if x.numeric == "Integer" and high == np.inf:
            raise ValueError(
                f"Dominio infinito no permitido para variable entera: {x.name}. "
                "Debe establecer un lmite superior finito para variables enteras."
            )

        # Ensure bounds are integers for range()
        if x.numeric == "Integer":
            low = int(low)
            high = int(high)

        domains.append(range(low, high+1))

    best_value = None
    best_assignment = None

    # 2. enumerar combinaciones
    for values in itertools.product(*domains):
        assignment = dict(zip([x() for x in variables], values))

        # 3. comprobar restricciones
        feasible = True
        for r in OP.restrictions:
            if not bool(r.subs(assignment)):
                feasible = False
                break

        if not feasible:
            continue

        # 4. evaluar objetivo
        value = OP.objective.subs(assignment)

        if best_value is None:
            best_value = value
            best_assignment = assignment
        else:
            if OP.kind == "max" and value > best_value:
                best_value = value
                best_assignment = assignment
            elif OP.kind == "min" and value < best_value:
                best_value = value
                best_assignment = assignment

    return best_assignment, best_value
```

# Valor Absoluto en la funcin objetivo

Un problema del tipo:
$$
\min: |f(x)|
$$

es **NO lineal**, y dificulta los solvers.
La idea es **reemplazar (|f(x)|)** por **variables auxiliares** que conviertan el problema en uno lineal.

* $( t^+ \ge 0 )$
* $( t^- \ge 0 )$

Y quiero hacer una __descomposicin__:

$$
z = |z| \quad \longrightarrow \quad z = t^+ - t^-, ; t^+, t^- \ge 0
$$

donde solo uno de ellos debe ser positivo y el otro cero.

Opciones:
1.  $( t^+ \cdot t^- = 0 )$
2.
   - $f(x) \le t^+$
   - $-f(x) \le t^-$
3. $(b*t^+ + (b-1)*t^-, b: Boolean$

de modo que automticamente uno de los dos se hace cero en el ptimo.

---

# **Cmo eliminar un valor absoluto en una restriccin**.

Si el problema original es:

$$
|g(x)| \le k
$$

Esto **no es lineal**, as que los solvers no lo entienden bien.

El valor absoluto se elimina separndolo en **dos restricciones lineales equivalentes**.

$$
|g(x)| \le k
\quad \Longleftrightarrow \quad
-k \le g(x) \le k
$$

Esto equivale a:

1. $(g(x) \le k)$
2. $(g(x) \ge -k)$

As el problema se vuelve **lineal y resoluble**.


Para resolver una restriccin con **valor absoluto del tipo **:

$$
|g(x)| \ge k
$$

Es equivalente a:

$$
g(x) \ge k \quad \text{o} \quad g(x) \le -k
$$

Es decir, **al menos una de las dos** debe cumplirse.

Esto es un **OR lgico**, y los OR **no pueden escribirse directamente** en programacin lineal.

Por eso aparece la tcnica del **Big-M + variables binarias** (modelado disyuntivo).

* $(b_1 \in Bool\{0,1\}) $
* $(b_2  \in Bool\{0,1\}) $

Y obliga a que **solo una de las dos sea 1**:

$$
b_1 + b_2 = 1
$$

Eso permite seleccionar **una de las dos ramas** del OR.

Ahora se usan restricciones con **Big-M**:

$$
g(x) \ge k - b_1 M
$$

Si (b_1 = 0), se obtiene:

$$
g(x) \ge k
$$

Si (b_1 = 1), la ecuacin queda:

$$
g(x) \ge k - M
$$

y como M es grande, la restriccin queda neutralizada.

solo hay dos opciones:

### **Opcin 1: b = 0, b = 1**

* Activa: $(g(x) \ge k)$
* Relaja: $(g(x) \le -k + M)$

### **Opcin 2: b = 1, b = 0**

* Activa: $(g(x) \le -k)$
* Relaja: $(g(x) \ge k - M)$

Por lo tanto, el modelo fuerza:

$$
g(x) \ge k \quad \text{o bien} \quad g(x) \le -k,
$$

que es EXACTAMENTE:

$$
|g(x)| \ge k.
$$


## Valor Absoluto En igualdad
$$
|g(x)| = k
$$
se transforma en

$$
b_1g(x) = b_1k
- b_2  g(x) =  b_2k \\
b_1 + b_2 = 1 \\
b_1, b_2 \in Bool
$$


## Tcnicas de modelizacin  
### Variables binarias

Las **variables binarias** permiten introducir **decisiones lgicas** dentro de un modelo de optimizacin lineal.  
Formalmente, una variable binaria toma valores en:

b  {0, 1}

y se interpreta como una respuesta **S / No**, **Activo / Inactivo**, **Elegido / No elegido**.

---

## Para qu sirven las variables binarias?

Las variables binarias permiten modelizar:

- Implicaciones lgicas entre decisiones
- Seleccin y activacin de unidades
- Restricciones condicionales (OR, AND)
- Valores excluyentes
- Costes fijos
- Aproximaciones lineales de funciones no lineales

La idea clave es siempre la misma:

> Convertir una condicin lgica en una decisin binaria auxiliar.

---

## Implicaciones entre variables binarias

Sean b, b  {0,1}.  
Las implicaciones lgicas pueden escribirse como restricciones lineales.

### Casos bsicos

- b = 0  b = 0  
   b  b

- b = 0  b = 1  
   b  1  b

- b = 1  b = 0  
   b  1  b

- b = 1  b = 1  
   b  b

Estas identidades permiten transformar **condicionales lgicos** en desigualdades lineales



## Activacin de una variable (on/off)

Sea x una variable real que solo puede tomar valor si b = 1.

Lb  x  Ub  
b  {0,1}

Interpretacin:

- Si b = 0  x = 0  
- Si b = 1  x  [L, U]

### Coste fijo de activacin

Si activar la variable tiene un coste C, se aade a la funcin objetivo:

Este patrn aparece en:
- apertura de plantas

- seleccin de proyectos
- costes hundidos

---

## Activacin de una restriccin (Big-M)

Sea una restriccin base:

 A x  k

Queremos que **solo se aplique si b = 1**.

### Forma tpica (Big-M)

 a x  k + M (1  b)  
y  {0,1}

Interpretacin:

- b = 1   a x  k  
- b = 0  restriccin relajada (k + M)

Anlogamente para restricciones :

$$
 a x  k  M (1  b)
$$

M debe ser **suficientemente grande**.

---

## OR lgico (disyuntivas)

Para imponer:

$$
|g(x)|  k  \\
 \\
(x)  k  \\
OR \\
 g(x)  k
$$

Se introducen binarias b, b  {0,1} con:

$$
b + b = 1
$$

y restricciones Big-M:

$$
g(x)  k  b M   \\
g(x)  k + b M
$$

De esta forma, **solo una rama del OR** puede activarse.

---

## Valor absoluto en igualdad

Para imponer:

$$
|g(x)| = k
$$

Se fuerza una de las dos igualdades:

$$
b  g(x) = b  k   \\
b  g(x) = b  k  
$$

con:

$$
b + b = 1    \\
b, b  \{0,1\}
$$
Esto convierte una condicin no lineal en un modelo lineal mixto.
